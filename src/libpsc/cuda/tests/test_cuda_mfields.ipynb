```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

import json
import numpy as np
```

```python
EX, EY, EZ, HX, HY, HZ = 3, 4, 5, 6, 7, 8

def as_field5d(dct):
    if '__field5d__' in dct:
        return np.array(dct['data']).swapaxes(2, 4)
    return dct

def cmflds_load(filename):
    file = open(filename)
    cmflds = json.load(file, object_hook=as_field5d)
    file.close()
    return cmflds

def plot_yz(cmflds, m, title=None):
    flds = cmflds["flds"]
    ib, im, dx = cmflds["ib"], cmflds["im"], cmflds["dx"]

    crds = [np.arange(ib[d] * dx[d], (ib[d]+im[d]) * dx[d], dx[d]) for d in range(3)]
    plt.figure()
    plt.pcolormesh(crds[1], crds[2], flds[0,m,0,:,:])
    plt.title(title)
```
